## Python for data science

Let's import dependence packages for our project.

```python
# Import dependence
import os    # for file path concatenation

import numpy as np   # for data preprocessing 
import pandas as pd  # for data preprocessing 

import matplotlib.pyplot as plt   # for data visualization
import seaborn as sns   # for data visualization
import plotly.express as px  # for data visualization
```

### Get data

Let's import our data on Breast Cancer from SKLEARN Python library.

```python
##########################################################################################
# Load Breast Cancer data sets.
##########################################################################################
# - The breast cancer dataset is a classic and very easy binary classification dataset.
# - (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html#sklearn.datasets.load_breast_cancer)
from sklearn import datasets

data = datasets.load_breast_cancer()
```

### Explore Data Analysis (EDA)

Once data is loaded, we start exploring the data and get insights and data quality issues. This process is sometimes referred to as Exploratory Data Analysis (EDA).

```python
# Determine features and label(s)
X = data['data']
y = data['target']
```

```python
# Check data
X[:5]
```

```python
y[:5]
```

```python
# Check data shape
# - Number of subjects (records) in the study.
X.shape
```

```python
y.shape
```

```python
# Check loaded data properties
data.keys()
```

```python
# Check target names
print(data['target_names'])
```

```python
# Check target names
print(data['feature_names'])
```

```python
##########################################################################################
# Make a Pandas data frame.
##########################################################################################
# Make a pandas data frame
df = pd.DataFrame(data=X, columns=data.feature_names)   # Add features first
df
```

```python
# Add labels to pandas data frame
df['target'] = pd.Series(y)

df
```

Once the data is transformed to a [Pandas data frame](https://pandas.pydata.org/), we check for any data quality issues. The first thing to check is for missing values, for example.

```python
# Check for missing values
# - No missing data found.
df.isnull().sum()
```

To get insights about each variable in your data set, I recommend using [Pandas Profiling](https://pandas-profiling.ydata.ai/docs/master/index.html) library. Pandas profiling generates a report with summary statistics and graphs for each variable from Pandas data frame. I show you here how to generate a profile report for Breast Cancer data set.

```python
# Pandas profile report for EDA
from pandas_profiling import ProfileReport
```

```python
# Generate profile report
profile_report = ProfileReport(df, minimal=True, title="Breast Cancer Data")
```

```python
# View profile report
profile_report.to_notebook_iframe()
```
